# Indexing methods in Pandas

**Indexing in pandas means simply selecting particular rows and columns of data from a DataFrame. Indexing could mean selecting all the rows and some of the columns, some of the rows and all of the columns, or some of each of the rows and columns. Indexing can also be known as Subset Selection.**

Link : https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html


```python
data["Age"][:5]
```

```python
data[["Age","Fare"]][:5]
```

```python
# iloc returns a Pandas Series when one row is selected, and a Pandas DataFrame when multiple rows are selected,
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html
data.iloc[0]
```

```python
data.iloc[0:5]
```

```python
data.iloc[0:5,1:3]
```

```python
# The loc property is used to access a group of rows and columns by label(s) or a boolean array. 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html
data.loc["Heikkinen, Miss. Laina"]
```

```python
data.loc["Heikkinen, Miss. Laina","Age"]
```

```python
data.loc["Heikkinen, Miss. Laina":"Allen, Mr. William Henry"]
```

```python
data.loc["Heikkinen, Miss. Laina":"Allen, Mr. William Henry",["Age","Fare"]]
```

```python
data.loc["Heikkinen, Miss. Laina","Age"] = 400
data.head()
```

```python
data [data["Age"]>30]
```

```python
data [data["Sex"]=="female"]
```

```python
data.loc[data["Sex"]=="male","Age"].mean()
```

```python
(data["Age"]>65).sum()
```

# Groupby in Pandas

**Pandas' GroupBy is a powerful and versatile function in Python. It allows you to split your data into separate groups to perform computations for better analysis.30**

Link : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html


```python
data_1 = pd.DataFrame({"K":["A","B","C","D","A","B","C","D"],
                      "data":[1,3,4,3,2,5,3,2]})
data_1
```

```python
data_1.groupby("K").sum()
```

```python
import numpy as np
data_1.groupby("K").aggregate(np.mean)
```
